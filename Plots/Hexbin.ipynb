```python
import pandas as pd
import numpy as np
import math
import kagglehub
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df_int = pd.read_csv('student_habits_performance_changed.csv')
```

```python
df_int.columns
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
axes = axes.flatten()
gs = 20

df_int.plot.hexbin(x='exam_score', y='study_hours_per_day', 
                   title='exam score by study hours (hexbin)', gridsize=gs,
                   ax=axes[0])

df_int.plot.hexbin(x='study_hours_per_day', y='exam_score',
                   title='study hours by exam score (hexbin)', gridsize=gs,
                   ax=axes[1], cmap='Purples')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
axes = axes.flatten()
df_int.plot.hexbin(x='social_media_hours', y='exam_score', 
                  title='Exam Score Hexbin Plot By Social Media Hours', gridsize=10,
                  cmap='viridis', ax=axes[0])
df_int.plot.hexbin(x='social_media_hours', y='study_hours_per_day',
                  gridsize=10, cmap='viridis', ax=axes[1],
                  title='Study Hours Per Day Hexbin Plot By Social Media Hours')
df_int.plot.hexbin(x='study_hours_per_day', y='exam_score', gridsize=10,
                  cmap='viridis', ax=axes[2],
                  title='Exam Score Hexbin Plot By Study Hours Per Day')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
axes = axes.flatten()
gs = 10
df_int.loc[
    df_int['gender']=='Male'
].plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (Male)', gridsize=gs,
              ax=axes[0], cmap='plasma')
df_int.loc[
    df_int['gender']=='Female'
].plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (Female)', gridsize=gs,
              ax=axes[1], cmap='plasma')
df_int.loc[
    df_int['gender']=='Other'
].plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (Other)', gridsize=gs,
              ax=axes[2], cmap='plasma')
df_int.plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (Total)', gridsize=gs,
              ax=axes[3], cmap='plasma')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 20))
axes = axes.flatten()
gs = 10
df_int.loc[
    df_int['mental_health_rating'] <= 4
].plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (Low Mental Health)', gridsize=gs,
              ax=axes[0], cmap='magma')
df_int.loc[
    df_int['mental_health_rating'].between(5, 8)
].plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (Medium Mental Health)', gridsize=gs,
              ax=axes[1], cmap='magma')
df_int.loc[
    df_int['mental_health_rating'] > 8
].plot.hexbin(x='study_hours_per_day', y='exam_score',
              title='study hours by exam score (High Mental Health)', gridsize=gs,
              ax=axes[2], cmap='magma')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))
gs = 10
df_int.plot.hexbin(x='social_media_hours', y='mental_health_rating',
              title='Mental Health Rating Hexbin Plot By Social Media Hours Per Day', gridsize=gs,
              ax=axes, cmap='cividis')
```
