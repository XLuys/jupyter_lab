```python
import pandas as pd
import math
```

```python
df = pd.read_csv('student_habits_performance_changed.csv')
```

```python
mean_value_study_hours_per_day = df['study_hours_per_day'].mean()
mean_value_exam_score = df['exam_score'].mean()
print(mean_value_study_hours_per_day, mean_value_exam_score)
```

```python
no_rows = df['study_hours_per_day'].shape[0] - 1
var_study_hours = 0
for index, value in df['study_hours_per_day'].items():
    var_study_hours += (value - mean_value_study_hours_per_day) ** 2
var_study_hours /= no_rows
st_dev_study_hours = math.sqrt(var_study_hours)
print(var_study_hours, st_dev_study_hours)
print(df['study_hours_per_day'].std())
```

```python
var_exam_score = 0
for index, value in df['exam_score'].items():

    var_exam_score += (value - mean_value_exam_score) ** 2
var_exam_score /= no_rows
std_exam_score = math.sqrt(var_exam_score)
print(var_exam_score, df['exam_score'].var())
print(std_exam_score, df['exam_score'].std())
```

```python
covariance = 0
for idx in df.index:
    covariance += (df['study_hours_per_day'].loc[idx] - mean_value_study_hours_per_day) * (df['exam_score'].loc[idx] - mean_value_exam_score)
covariance /= no_rows
print(covariance)
```

```python
cov = df['exam_score'].cov(df['study_hours_per_day'])
print(cov)
```

```python
corr = cov/(st_dev_study_hours * std_exam_score)
corr_std = df['exam_score'].corr(df['study_hours_per_day'])
print(corr, corr_std)
```
